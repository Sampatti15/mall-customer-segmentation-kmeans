Selecting the Annual Income and Spending Score columns 

```python
X = customer_data.iloc[:,[3,4]].values
print(X)
```

Selecting the number of clusters

WCSS = Within Clusters Sum of squares

```python
# Finding WCSS value for different number of clusters 

wcss = []

for i in range(1,11) : 
    Kmeans = KMeans (n_clusters=i, init='k-means++', random_state=42)
    Kmeans.fit(X)

    wcss.append(Kmeans.inertia_)
```

```python
# Plotting an elbow graph

sns.set()
plt.plot(range(1,11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

Optimum number of clusters => 5


Training the K-Means Clustering Model

```python
Kmeans = KMeans(n_clusters=5, init='k-means++', random_state=0)

# Return label for each data point on their cluster
Y = Kmeans.fit_predict(X)     # Fit predict will return cluster no.for each data point
print(Y)  
```

Visualizing the clusters

```python
# Plotting all the clusters and their centroids 

plt.figure(figsize=(7,7))
plt.scatter(X[Y==0,0], X[Y==0,1], s=40, c='red', label='Cluster 1')
plt.scatter(X[Y==1,0], X[Y==1,1], s=40, c='yellow', label='Cluster 2')
plt.scatter(X[Y==2,0], X[Y==2,1], s=40, c='green', label='Cluster 3')
plt.scatter(X[Y==3,0], X[Y==3,1], s=40, c='blue', label='Cluster 4')
plt.scatter(X[Y==4,0], X[Y==4,1], s=40, c='brown', label='Cluster 5')

# Plotting the centroids 
plt.scatter(Kmeans.cluster_centers_[:,0], Kmeans.cluster_centers_[:,1], c='cyan', label='centroids')

plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
